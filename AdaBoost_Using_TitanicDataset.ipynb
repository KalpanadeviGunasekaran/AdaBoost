```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
train=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
train
```

```python
train.info(),test.info()
```

```python
#set the PassengerId column as index
train.set_index("PassengerId",inplace=True)
test.set_index("PassengerId",inplace=True)
```

```python
train
```

```python
y_train=train["Survived"]
```

```python
train.drop(labels="Survived",axis=1,inplace=True)
```

```python
train.shape,test.shape
```

```python
train_test=train.append(test)
```

```python
coldel=["Name","Age","SibSp","Ticket","Cabin","Parch","Embarked"]
train_test.drop(labels=coldel,axis=1,inplace=True)
```

```python
train_test
```

```python
#one_hot_encoding convert obj into numberical
train_test_dummies=pd.get_dummies(train_test,columns=["Sex"])
```

```python
train_test_dummies.shape
```

```python
train_test_dummies.info()
```

```python
train_test_dummies.isnull().sum()
```

```python
train_test_dummies.fillna(value=0.0,inplace=True)
train_test_dummies.isnull().sum()
```

```python
X_train=train_test_dummies.values[0:891]
X_test=train_test_dummies.values[891:]
```

```python
from sklearn.preprocessing import StandardScaler
model=StandardScaler()
X_train_scale=model.fit_transform(X_train)
X_test_scale=model.transform(X_test)
```

```python
from sklearn.model_selection import train_test_split
X_train_sub,X_validation_sub,y_train_sub,y_validation_sub=train_test_split(X_train_scale,y_train,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
base_classifier = DecisionTreeClassifier(max_depth=3)
adaboost_classifier = AdaBoostClassifier(base_classifier, n_estimators=50, random_state=42)
adaboost_classifier.fit(X_train_sub, y_train_sub)
y_pred = adaboost_classifier.predict(X_validation_sub)
accuracy = accuracy_score(y_validation_sub, y_pred)
print(f"Accuracy: {accuracy}")
```
